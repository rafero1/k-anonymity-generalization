# Tratamento do csv de regiões
Para ajudar na generalização de municípios

```python
import pandas as pd

df = pd.read_csv('ce-regions.csv', header=0,
                 names=['municipioCaso', 'regiaoPlanejamentoCaso'])
df
```

Transformar para UPPERCASE e remover acentuação

```python
df["municipioCaso"] = df["municipioCaso"].str.upper()
df['municipioCaso'] = df['municipioCaso'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df["regiaoPlanejamentoCaso"] = df["regiaoPlanejamentoCaso"].str.upper()
df["regiaoPlanejamentoCaso"] = df['regiaoPlanejamentoCaso'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df
```

```python
df.to_csv('ce-regions-format.csv', index=False)
```
